# Here In This project we Analyzing IPL data using Python .

```python
# importing all required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the data that is in csv file
ipl = pd.read_csv('matches.csv')
```

```python
ipl.head()
```

```python
# looking for data shape rows, columns
ipl.shape
```

```python
# geeting the frequency of most man of the match award
ipl['player_of_match'].value_counts()
```

```python
# getting the top 10 players of list with man of the match
ipl['player_of_match'].value_counts()[0:10]
```

```python
#getting the top 5 players list with man of the match awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
# to get only keys or names of top five players
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
# making a bar plot of these 5 playrs with most man of the match awards
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color='g')
plt.title('Most Man of the Match awarded player')
plt.show()
```

```python
# Now getting the frequency of result column
ipl['result'].value_counts()
```

```python
# finding out the team who won most of the toss
ipl['toss_winner'].value_counts()
```

```python
# how many team won the match by doing first match
batting_first_won = ipl[ipl['win_by_runs']!=0]
```

```python
# looking the head column
batting_first_won.head()
```

```python
# making the histogram 
plt.figure(figsize=(7,7))
plt.hist(batting_first_won['win_by_runs'])
plt.title('Matches won by Runs')
plt.xlabel('Runs')
plt.ylabel('Matches')
plt.show()
```

```python
# finding out the number of wins with respect to, each team after batting first
batting_first_won['winner'].value_counts()
```

```python
# making a bar plot of 4 most winning team after batting first
plt.figure(figsize=(8,8))
plt.bar(list(batting_first_won['winner'].value_counts()[0:4].keys()),list(batting_first_won['winner'].value_counts()[0:4]),color=['blue','yellow','red','purple'])
plt.title('Team who won most matches by doing batting first')
plt.show()
```

```python
# making the pie chart of same data that is most won matches by doing batting first
plt.figure(figsize=(10,10))
plt.pie(list(batting_first_won['winner'].value_counts()),labels=list(batting_first_won['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title('Pie chart distribution of matches won by teams doing batting first')
plt.show()
```

```python
# extracting thoses records where a team won after doing batting second
batting_second_won = ipl[ipl['win_by_wickets']!=0]
```

```python
# looking at top 5 data
batting_second_won.head()
```

```python
# making a histogram for frequency where a team has won after doing seconf batting
plt.figure(figsize=(8,8))
plt.hist(batting_second_won['win_by_wickets'],bins=30)
plt.title('Histogram of those teams who won after doing second batting')
plt.xlabel('won by wickets')
plt.ylabel('matches')
plt.show()
```

```python
# finding the frequency of number of wins with respect to each team after doing second batting
batting_second_won['winner'].value_counts()
```

```python
# making a bar plot for top 4 wins of each time after doing batting  second
plt.figure(figsize=(10,10))
plt.bar(list(batting_second_won['winner'].value_counts()[0:4].keys()),list(batting_second_won['winner'].value_counts()[0:4]),color=['purple','blue','red','navy'])
plt.title('Team who wonn most matches by doing second batting')
plt.show()                                                                                                                                    
```

```python
# making a pie chart for distributution of most wins after batting second
plt.figure(figsize=(12,12))
plt.pie(list(batting_second_won['winner'].value_counts()),labels=list(batting_second_won['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title('Pie chart distribution of most wins after second batting')
plt.show()
```

```python
#Looking at the numbers of matches played at each season
ipl['season'].value_counts()
```

```python
# making a histo gram of most matches according to season
plt.figure(figsize=(8,8))
plt.hist(ipl['season'],bins=30)
plt.title('Most played matches according to seasons')
plt.xlabel('IPL Year')
plt.ylabel('Total Played Matches in one year ')
plt.show()
```

```python
# looking at the numbersof matches played in each city
ipl['city'].value_counts()
```

```python
# maaking a pie chart of most numbers of matches played in which city
plt.figure(figsize=(12,12))
plt.pie(list(ipl['city'].value_counts()),labels=list(ipl['city'].value_counts().keys()),autopct='%0.1f%%')
plt.title('Maximum number of matches played in city')
plt.show()
```
